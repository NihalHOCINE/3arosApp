```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming 'my_total_data' is your DataFrame

# Convert 'Date' column to datetime type if it's not already
my_total_data_stock_normal_effet_marge['Date'] = pd.to_datetime(my_total_data['Date'])

# Extract month and year from the 'Date' column
my_total_data_stock_normal_effet_marge['Month'] = my_total_data_stock_normal_effet_marge['Date'].dt.to_period('M')
my_total_data_stock_normal_effet_marge['Year'] = my_total_data_stock_normal_effet_marge['Date'].dt.to_period('Y')

# Group by month and calculate the sum of 'Effet_Marge'
monthly_aggregated = my_total_data_stock_normal_effet_marge.groupby('Month')['Effet_Marge'].sum()

# Group by year and calculate the sum of 'Effet_Marge'
yearly_aggregated = my_total_data_stock_normal_effet_marge.groupby('Year')['Effet_Marge'].sum()

# Plotting for monthly aggregation
plt.figure(figsize=(12, 6))
monthly_aggregated.plot(marker='o', linestyle='-', label='Monthly Aggregation')
plt.title('Effet Marge Over Time (Monthly Aggregation)')
plt.xlabel('Month')
plt.ylabel('Total Effet Marge')
plt.legend()
plt.grid(True)
plt.show()

# Plotting for yearly aggregation
plt.figure(figsize=(12, 6))
yearly_aggregated.plot(marker='o', linestyle='-', label='Yearly Aggregation')
plt.title('Effet Marge Over Time (Yearly Aggregation)')
plt.xlabel('Year')
plt.ylabel('Total Effet Marge')
plt.legend()
plt.grid(True)
plt.show()
```
