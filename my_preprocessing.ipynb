```python
#fill missing values in 'Reference Produit' based on the mode (most frequent value) for each 'FamilleArticle'

# Group by 'FamilleArticle' and find the mode for each group
reference_mode = my_total_data.groupby('FamilleArticle')['Reference Produit'].apply(lambda x: x.mode().iloc[0] if not x.mode().empty else None)

# Fill missing values based on the mode of the corresponding 'FamilleArticle'
my_total_data['Reference Produit'] = my_total_data.apply(
    lambda row: reference_mode[row['FamilleArticle']] if pd.isnull(row['Reference Produit']) else row['Reference Produit'],
    axis=1
)

# Display the result
print(my_total_data[['FamilleArticle', 'Reference Produit']])
```

```python
# the best sales depending on the season ( avec mantants de bon de livraison)

import pandas as pd
import matplotlib.pyplot as plt


# Define a function to determine the season based on the month
def get_season(month):
    if 3 <= month <= 5:
        return 'Spring'
    elif 6 <= month <= 8:
        return 'Summer'
    elif 9 <= month <= 11:
        return 'Fall'
    else:
        return 'Winter'

# Create a new column 'Season' based on the 'Date' column
my_total_data['Season'] = my_total_data['Date'].dt.month.apply(get_season)

# Filter for 'Bon de livraison'
livraison_data = my_total_data[my_total_data['TypeEffet'] == 'Bon de livraison']

# Group by 'Season' and calculate the sum of MontantHT
seasonal_sales = livraison_data.groupby('Season')['MontantHT'].sum()

# Plot the results
seasonal_sales.plot(kind='bar', color='skyblue')
plt.title('Seasonal Sales Comparison')
plt.xlabel('Season')
plt.ylabel('Total Sales (MontantHT)')
plt.show()
```

```python
#how many function : check if montant of a product in bon dentrée is the same in de livraison + calculate the difference 
import pandas as pd

def calculate_montant_changes(data):
    # Sort the DataFrame by 'Date'
    data = data.sort_values(by='Date')

    # Create variables to count changes
    same_count = 0
    changed_count = 0

    # Create a mask for 'Bon de livraison'
    livraison_mask = data['TypeEffet'] == 'Bon de livraison'

    # Iterate through the rows using iterrows
    for index, row in data[livraison_mask].iterrows():
        item_condition = (
            (data['Reference Produit'] == row['Reference Produit']) &
            (data['FamilleArticle'] == row['FamilleArticle']) &
            (data['Code Produit'] == row['Code Produit'])
        )

        # Check for previous entry with 'TypeEffet' equal to 'Bon d'entrée'
        previous_entree = data.loc[item_condition & (data['TypeEffet'] == "Bon d'entrée") & (data.index < index)]

        if not previous_entree.empty:
            # Calculate the difference in MontantHT
            diff_amount = row['MontantHT'] - previous_entree.iloc[-1]['MontantHT']

            if diff_amount == 0:
                same_count += 1
            else:
                changed_count += 1

    return same_count, changed_count

# Call the function with your DataFrame
same_count, changed_count = calculate_montant_changes(my_total_data)

# Print the results
print(f"Number of entries where MontantHT stayed the same: {same_count}")
print(f"Number of entries where MontantHT changed: {changed_count}")
```
